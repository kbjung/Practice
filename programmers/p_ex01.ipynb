# 성격 유형 검사하기
- survey의 원소는 "RT", "TR", "FC", "CF", "MJ", "JM", "AN", "NA" 중 하나
- 성격유형

  지표 번호 | 성격 유형
  --|--
  1번 지표 | 라이언형(R), 튜브형(T)
  2번 지표 | 콘형(C), 프로도형(F)
  3번 지표 | 제이지형(J), 무지형(M)
  4번 지표 | 어피치형(A), 네오형(N)

- 질문에 대한 성격 유형 점수 예시

  선택지| 성격 유형 점수
  --|--
  매우 비동의 | 네오형 3점
  비동의 | 네오형 2점
  약간 비동의 | 네오형 1점
  모르겠음 | 어떤 성격 유형도 점수를 얻지 않습니다
  약간 동의 | 어피치형 1점
  동의 | 어피치형 2점
  매우 동의 | 어피치형 3점

- choices 설명

  choices | 뜻
  --|--
  1 | 매우 비동의
  2 | 비동의
  3 | 약간 비동의
  4 | 모르겠음
  5 | 약간 동의
  6 | 동의
  7 | 매우 동의

  </br>
- 예시

  survey | choices | result
  --|--|--
  ["AN", "CF", "MJ", "RT", "NA"] | [5, 3, 2, 7, 5] | "TCMA"
  ["TR", "RT", "TR"] | [7, 1, 3] | "RCJA"
  
  </br>  

  - 1번 예시 설명  
  </br>
  
  지표 번호 | 성격 유형 | 점수 | 성격 유형 | 점수
  --|--|--|--|--
  1번 지표 | 라이언형(R) | 0 | 튜브형(T) | 3
  2번 지표 | 콘형(C)     | 1 | 프로도형(F) | 0
  3번 지표 | 제이지형(J) | 0 | 무지형(M) | 2
  4번 지표 | 어피치형(A) | 1 | 네오형(N) | 1

  각 지표에서 더 점수가 높은 T,C,M이 성격 유형입니다.  
  하지만, 4번 지표는 1점으로 동일한 점수입니다. 따라서, 4번 지표의 성격 유형은 사전순으로 빠른 A입니다.  

  따라서 "TCMA"를 return 해야 합니다.  

  - 2번 예시 설명  
  </br>

  지표 번호 | 성격 유형 | 점수 | 성격 유형 | 점수
  --|--|--|--|--
  1번 지표 | 라이언형(R) | 6 | 튜브형(T) | 1
  2번 지표 | 콘형(C)     | 0 | 프로도형(F) | 0
  3번 지표 | 제이지형(J) | 0 | 무지형(M) | 0
  4번 지표 | 어피치형(A) | 0 | 네오형(N) | 0

  1번 지표는 튜브형(T)보다 라이언형(R)의 점수가 더 높습니다. 따라서 첫 번째 지표의 성격 유형은 R입니다.  
  하지만, 2, 3, 4번 지표는 모두 0점으로 동일한 점수입니다. 따라서 2, 3, 4번 지표의 성격 유형은 사전순으로 빠른 C, J, A입니다.  

  따라서 "RCJA"를 return 해야 합니다.

```python
import pandas as pd
```

```python
def solution(survey, choices):
    ch_df = pd.DataFrame({'ch':['RT', 'CF', 'JM', 'AN']})
    df_list = []
    for i in range(len(survey)):
        df_list.append(0)
    dict = {"R":df_list, "T":df_list, "F":df_list, "C":df_list, "M":df_list, "J":df_list, "A":df_list, "N":df_list, "ch":df_list}
    df = pd.DataFrame(dict)

    for i in range(len(survey)):
        if survey[i] == 'RT':
            df.loc[i, 'ch'] = "RT"
            if choices[i] == 1:
                df.loc[i, 'R'] = 3
            elif choices[i] == 2:
                df.loc[i, 'R'] = 2
            elif choices[i] == 3:
                df.loc[i, 'R'] = 1
            elif choices[i] == 5:
                df.loc[i, 'T'] = 1
            elif choices[i] == 6:
                df.loc[i, 'T'] = 2
            elif choices[i] == 7:
                df.loc[i, 'T'] = 3
        elif survey[i] == 'TR':
            df.loc[i, 'ch'] = "TR"
            if choices[i] == 1:
                df.loc[i, 'T'] = 3
            elif choices[i] == 2:
                df.loc[i, 'T'] = 2
            elif choices[i] == 3:
                df.loc[i, 'T'] = 1
            elif choices[i] == 5:
                df.loc[i, 'R'] = 1
            elif choices[i] == 6:
                df.loc[i, 'R'] = 2
            elif choices[i] == 7:
                df.loc[i, 'R'] = 3

        elif survey[i] == 'FC':
            df.loc[i, 'ch'] = "FC"
            if choices[i] == 1:
                df.loc[i, 'F'] = 3
            elif choices[i] == 2:
                df.loc[i, 'F'] = 2
            elif choices[i] == 3:
                df.loc[i, 'F'] = 1
            elif choices[i] == 5:
                df.loc[i, 'C'] = 1
            elif choices[i] == 6:
                df.loc[i, 'C'] = 2
            elif choices[i] == 7:
                df.loc[i, 'C'] = 3
        elif survey[i] == 'CF':
            df.loc[i, 'ch'] = "CF"
            if choices[i] == 1:
                df.loc[i, 'C'] = 3
            elif choices[i] == 2:
                df.loc[i, 'C'] = 2
            elif choices[i] == 3:
                df.loc[i, 'C'] = 1
            elif choices[i] == 5:
                df.loc[i, 'F'] = 1
            elif choices[i] == 6:
                df.loc[i, 'F'] = 2
            elif choices[i] == 7:
                df.loc[i, 'F'] = 3

        elif survey[i] == 'MJ':
            df.loc[i, 'ch'] = "MJ"
            if choices[i] == 1:
                df.loc[i, 'M'] = 3
            elif choices[i] == 2:
                df.loc[i, 'M'] = 2
            elif choices[i] == 3:
                df.loc[i, 'M'] = 1
            elif choices[i] == 5:
                df.loc[i, 'J'] = 1
            elif choices[i] == 6:
                df.loc[i, 'J'] = 2
            elif choices[i] == 7:
                df.loc[i, 'J'] = 3
        elif survey[i] == 'JM':
            df.loc[i, 'ch'] = "JM"
            if choices[i] == 1:
                df.loc[i, 'J'] = 3
            elif choices[i] == 2:
                df.loc[i, 'J'] = 2
            elif choices[i] == 3:
                df.loc[i, 'J'] = 1
            elif choices[i] == 5:
                df.loc[i, 'M'] = 1
            elif choices[i] == 6:
                df.loc[i, 'M'] = 2
            elif choices[i] == 7:
                df.loc[i, 'M'] = 3

        elif survey[i] == 'AN':
            df.loc[i, 'ch'] = "AN"
            if choices[i] == 1:
                df.loc[i, 'A'] = 3
            elif choices[i] == 2:
                df.loc[i, 'A'] = 2
            elif choices[i] == 3:
                df.loc[i, 'A'] = 1
            elif choices[i] == 5:
                df.loc[i, 'N'] = 1
            elif choices[i] == 6:
                df.loc[i, 'N'] = 2
                df.loc[i, 'ch'] = "AN"
            elif choices[i] == 7:
                df.loc[i, 'N'] = 3
        elif survey[i] == 'NA':
            df.loc[i, 'ch'] = "NA"
            if choices[i] == 1:
                df.loc[i, 'N'] = 3
            elif choices[i] == 2:
                df.loc[i, 'N'] = 2
            elif choices[i] == 3:
                df.loc[i, 'N'] = 1
            elif choices[i] == 5:
                df.loc[i, 'A'] = 1
                df.loc[i, 'ch'] = "NA"
            elif choices[i] == 6:
                df.loc[i, 'A'] = 2
            elif choices[i] == 7:
                df.loc[i, 'A'] = 3
    
    df = df.replace({'ch':{'TR':'RT', 'FC':'CF', 'MJ':'JM', 'NA':'AN'}})

    try:
        temp = df[df['ch'] == 'RT']
        idx_list = temp.index
        temp_sum = temp.sum(numeric_only=True)
        df.loc[idx_list[0], 'R'] = temp_sum['R']
        df.loc[idx_list[0], 'T'] = temp_sum['T']
    except:
        pass

    try:
        temp = df[df['ch'] == 'CF']
        idx_list = temp.index
        temp_sum = temp.sum(numeric_only=True)
        df.loc[idx_list[0], 'F'] = temp_sum['F']
        df.loc[idx_list[0], 'C'] = temp_sum['C']
    except:
        pass

    try:
        temp = df[df['ch'] == 'JM']
        idx_list = temp.index
        temp_sum = temp.sum(numeric_only=True)
        df.loc[idx_list[0], 'M'] = temp_sum['M']
        df.loc[idx_list[0], 'J'] = temp_sum['J']
    except:
        pass

    try:
        temp = df[df['ch'] == 'AN']
        idx_list = temp.index
        temp_sum = temp.sum(numeric_only=True)
        df.loc[idx_list[0], 'A'] = temp_sum['A']
        df.loc[idx_list[0], 'N'] = temp_sum['N']
    except:
        pass
    
    df = df.drop_duplicates('ch')
    df = pd.merge(ch_df, df, on='ch', how='left').fillna(0)

    answer = ''
    ch_list = df['ch'].to_list()
    for i, one in enumerate(ch_list):
        s1, s2 = sorted(one)
        if df.loc[i, s1] >= df.loc[i, s2]:
            answer += s1
        else:
            answer += s2
    return answer
```

```python
def solution(survey, choices):
    answer = ''
    r, t, c, f, j, m, a, n = 0, 0, 0, 0, 0, 0, 0, 0
    for i, one in enumerate(survey):
        if one == 'RT':
            if choices[i] == 1:
                r += 3
            elif choices[i] == 2:
                r += 2
            elif choices[i] == 3:
                r += 1
            elif choices[i] == 5:
                t += 1
            elif choices[i] == 6:
                t += 2
            elif choices[i] == 7:
                t += 3
        elif one == 'TR':
            if choices[i] == 1:
                t += 3
            elif choices[i] == 2:
                t += 2
            elif choices[i] == 3:
                t += 1
            elif choices[i] == 5:
                r += 1
            elif choices[i] == 6:
                r += 2
            elif choices[i] == 7:
                r += 3

        elif one == 'FC':
            if choices[i] == 1:
                f += 3
            elif choices[i] == 2:
                f += 2
            elif choices[i] == 3:
                f += 1
            elif choices[i] == 5:
                c += 1
            elif choices[i] == 6:
                c += 2
            elif choices[i] == 7:
                c += 3
        elif one == 'CF':
            if choices[i] == 1:
                c += 3
            elif choices[i] == 2:
                c += 2
            elif choices[i] == 3:
                c += 1
            elif choices[i] == 5:
                f += 1
            elif choices[i] == 6:
                f += 2
            elif choices[i] == 7:
                f += 3

        elif one == 'JM':
            if choices[i] == 1:
                j += 3
            elif choices[i] == 2:
                j += 2
            elif choices[i] == 3:
                j += 1
            elif choices[i] == 5:
                m += 1
            elif choices[i] == 6:
                m += 2
            elif choices[i] == 7:
                m += 3
        elif one == 'MJ':
            if choices[i] == 1:
                m += 3
            elif choices[i] == 2:
                m += 2
            elif choices[i] == 3:
                m += 1
            elif choices[i] == 5:
                j += 1
            elif choices[i] == 6:
                j += 2
            elif choices[i] == 7:
                j += 3

        elif one == 'AN':
            if choices[i] == 1:
                a += 3
            elif choices[i] == 2:
                a += 2
            elif choices[i] == 3:
                a += 1
            elif choices[i] == 5:
                n += 1
            elif choices[i] == 6:
                n += 2
            elif choices[i] == 7:
                n += 3
        elif one == 'NA':
            if choices[i] == 1:
                n += 3
            elif choices[i] == 2:
                n += 2
            elif choices[i] == 3:
                n += 1
            elif choices[i] == 5:
                a += 1
            elif choices[i] == 6:
                a += 2
            elif choices[i] == 7:
                a += 3

    if r >= t:
        answer += 'R'
    else:
        answer += 'T'
    if c >= f:
        answer += 'C'
    else:
        answer += 'F'
    if j >= m:
        answer += 'J'
    else:
        answer += 'M'
    if a >= n:
        answer += 'A'
    else:
        answer += 'N'

    return answer
```

## 기본 검사 점수 데이터 셋 만들기
